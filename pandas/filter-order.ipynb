Filtering and Ordering 

```python
import pandas as pd 
```

```python
df = pd.read_csv('world_population.csv')  # Load data from a CSV file into a DataFrame

print(df)
```

```python
ranked_countries = df[df['Rank'] > 100] # Filter the DataFrame to include only countries with a rank greater than 100
print(ranked_countries)
```

```python
specific_countries = df[df['Country/Territory'].isin(['China', 'India', 'United States'])] # Filter the DataFrame to include only specific countries
print(specific_countries)
```

```python
str_include_countries = df[df['Country/Territory'].str.contains('United')] # Filter the DataFrame to include only countries whose names contain 'United'
print(str_include_countries)
```

```python
df2 = df.set_index('Country/Territory')  # Set the 'Country/Territory' column as the index of the DataFrame

#print(df2) # Print the DataFrame with the new index

row_data = df2.filter(items=['China', 'India', 'United States'] , axis=0)  # Filter the DataFrame to include only specific countries using the new index
#print(row_data)

column_data = df2.filter(items=['Rank', '2020 Population'], axis=1)  # Filter the DataFrame to include only specific columns
#print(column_data)

similar_columns = df2.filter(like='2020', axis=1)  # Filter the DataFrame to include only columns that contain '2020' in their names
print(similar_columns)

similar_rows = df2.filter(like='United', axis=0)  # Filter the DataFrame to include only rows that contain 'United' in their index  
print(similar_rows)
```

Sorting


```python
sorted_df = df.sort_values(by='2020 Population', ascending=False)  # Sort the DataFrame by the '2020 Population' column in descending order
print(sorted_df)
```
