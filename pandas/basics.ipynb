```python
import pandas as pd
```

| File Extension    | Pandas Function                         |
| ----------------- | --------------------------------------- |
| `.csv`            | `pd.read_csv()`                         |
| `.xlsx`, `.xls`   | `pd.read_excel()`                       |
| `.json`           | `pd.read_json()`                        |
| `.parquet`        | `pd.read_parquet()`                     |
| `.feather`        | `pd.read_feather()`                     |
| `.pickle`, `.pkl` | `pd.read_pickle()`                      |
| `.html`           | `pd.read_html()`                        |
| `.xml`            | `pd.read_xml()`                         |
| `.sql`            | `pd.read_sql()` / `pd.read_sql_query()` |
| `.sas7bdat`       | `pd.read_sas()`                         |
| `.dta` (Stata)    | `pd.read_stata()`                       |
| `.h5`, `.hdf5`    | `pd.read_hdf()`                         |
| `.orc`            | `pd.read_orc()`                         |
