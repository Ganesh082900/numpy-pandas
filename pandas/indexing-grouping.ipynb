Indexing


```python
import pandas as pd
```

```python
df = pd.read_csv('world_population.csv')  # Load data from a CSV file into a DataFrame
print(df)  # Display the DataFrame
```

```python
#print(df.info())
print(df.columns.tolist())

df.set_index('Continent', inplace=True)  # Set 'Continent' column as the index

df.sort_index()  # Sort the DataFrame by index (Continent)
```

Grouping

```python
grouped = df.groupby('2022 Population')  # Group the DataFrame by '2022 Population'

#grouped.get_group('Asia')  # Get the group corresponding to 'Asia'

grouped.mean()  # Calculate the mean of each group
```

```python
# To avoid the above error, we can use the numeric_only= True parameter in the mean() function.
grouped.mean(numeric_only=True)
```
