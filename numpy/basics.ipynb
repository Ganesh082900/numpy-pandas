Time taken for the same task in list and np_array

```python
import time

start = time.time()
py_list = [i*2 for i in range(100000000)]
end = time.time()
print(f"Python list creation time: {end - start}")

start = time.time()
np_array = np.arange(100000000) * 2
end = time.time()
print(f"NumPy array creation time: {end - start}")
```

Vector , Matrix and Tensor

```python
# vector 

vector = np.array([1, 2, 3, 4, 5])
print("Vector:\n", vector)

# matrix

matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Matrix:\n", matrix)

# Tensor

tensor = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("Tensor:\n", tensor)
```

Create an array with Dtype 

```python
arr_float = np.array([1, 2, 3, 4, 5], dtype=np.float32)   
print("Array with specified data type:\n", arr_float)
print("Data type of the array:", arr_float.dtype)


arr_int = np.array([1, 2, 3, 4, 5], dtype=np.int32)
print("Array with specified data type:\n", arr_int)
print("Data type of the array:", arr_int.dtype)

arr_complex = np.array([1, 2, 3, 4, 5], dtype=np.complex64)
print("Array with specified data type:\n", arr_complex) 
print("Data type of the array:", arr_complex.dtype)
```
