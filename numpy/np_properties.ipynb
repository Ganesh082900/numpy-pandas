```python
import numpy as np
```

Array properties

```python
arr = np.array([[1, 45, 32], [28, 7, 76], [23, 56, 89], [12, 34, 56]])
print("Array:\n", arr)
print("Shape of the array:", arr.shape) # gives the shape of the array (number of rows, number of columns)
print("Number of dimensions:", arr.ndim) # gives the number of dimensions of the array
print("Size of the array:", arr.size) # gives the total number of elements in the array
print("Data type of the array:", arr.dtype) # gives the data type of the array
print("Item size (in bytes):", arr.itemsize) # gives the size of each element in the array
print("Total bytes consumed by the array:", arr.nbytes) # gives the total bytes consumed by the elements of the array
print("Transpose of the array:\n", arr.T) # gives the transpose of the array
```

Array Reshaping

```python
arr = np.arange(10)
print("Original array:", arr)

reshaped_arr = arr.reshape(2, 5) # Reshape the array to 2 rows and 5 columns
print("Reshaped array (2x5):\n", reshaped_arr)

flattened_arr = reshaped_arr.flatten() # Flatten the array back to 1D - returns a copy of the array collapsed into one dimension
print("Flattened array:", flattened_arr)

raveled_arr = reshaped_arr.ravel() # Ravel the array back to 1D (returns a flattened view) - returns a flattened view of the array, which is a reference to the original array whenever possible
print("Raveled array:", raveled_arr)

reversed_arr = arr[::-1] # Reverse the array
print("Reversed array:", reversed_arr)
```

NUMPY properties 

```python
# Zero - By default the dtype of the array is float64

np_zeros = np.zeros((3, 4))
print(np_zeros)
```

```python
# Ones

np_ones = np.ones((2, 3))
print(np_ones)
```

```python
# full - Create an array filled with a specific value
# parameters: shape, fill_value

np_full = np.full((2, 3), 5)
print(np_full)
```

```python
# random.rand - Create an array with random values
# parameters: shape

np_random = np.random.rand(2, 3)
print(np_random)

# random.uniform - Create an array with random float values from a uniform distribution
# parameters: low, high, size

np_uniform = np.random.uniform(1, 10, size=(2, 3))
print(np_uniform)

# random.randint - Create an array with random integer values
# parameters: low, high, size

np_randint = np.random.randint(1, 10, size=(2, 3))
print(np_randint)
```

```python
# arrange - Create an array with evenly spaced values within a given interval
# parameters: start, stop, step(intervals between values)

np_arange = np.arange(0, 10, 2)
print(np_arange)
```

Generate large data set using random and pcg

```python
from numpy.random import Generator as gen, PCG64 as pcg

arr_rg = gen(pcg(seed=42)) # Create a random number generator with a specific seed for reproducibility

arr_rg.normal(size=(2, 3)) # Create an array with random float values from a normal distribution
```

```python
arr_rg = gen(pcg(seed=42)) # Create a random number generator with a specific seed for reproducibility

arr_rg.integers(1, 10, size=(2, 3)) # Create an array with random integer values between 1 and 10
```

```python
arr_rg.random(size=(2, 3)) # Create an array with random float values from a normal distribution
```
