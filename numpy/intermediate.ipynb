```python
import numpy as np
import matplotlib.pyplot as plt
```

vector and matrix Additions|Multiplication

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([10, 20, 30, 40, 50])

vector_sum = vector1 + vector2
print("Vector Sum:", vector_sum)

vector_product = vector1 * vector2
print("Vector Product:", vector_product)

vector_dot_product = np.dot(vector1, vector2)
print("Vector Dot Product:", vector_dot_product)

```

```python
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

matrix_sum = matrix1 + matrix2
print("Matrix Sum:\n", matrix_sum)

matrix_product = np.dot(matrix1, matrix2)
print("Matrix Product:\n", matrix_product)

matrix_dot_product = np.dot(matrix1, matrix2)
print("Matrix Dot Product:\n", matrix_dot_product)
```

```python
# Data structure: [company sales ID, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Delta Corp
    [2, 120000, 140000, 160000, 190000],  # Beta Ltd
    [3, 200000, 230000, 260000, 300000],  # Gamma Inc
    [4, 180000, 210000, 240000, 270000],  # Delta Corp
    [5, 160000, 185000, 205000, 230000]   # Epsilon Co
])
```

```python
print("Sales Data:\n", sales_data)
print("Shape of Sales Data:", sales_data.shape)
# 1st 3 companies sales data
subset_sales_data = sales_data[:3]
print("Subset of Sales Data (1st 3 companies):\n", subset_sales_data)
```

```python
# Total sales of the years 2021 to 2024 for each company
total_sales = np.sum(sales_data[:, 1:], axis=1)
print("Total Sales for each company from 2021 to 2024:\n", total_sales)
```

```python
# Total sales of each company till date
total_sales_till_date = np.sum(sales_data[:, 1:], axis=0)
print("Total Sales for each company till date:\n", total_sales_till_date)
```

```python
# min and max sales for each company
min_sales = np.min(sales_data[:, 1:], axis=1)
print("Minimum Sales for each company from 2021 to 2024:\n", min_sales)
max_sales = np.max(sales_data[:, 1:], axis=1)
print("Maximum Sales for each company from 2021 to 2024:\n", max_sales)
```

```python
#min and max sales for each year
min_sales_year = np.min(sales_data[:, 1:], axis=0)
print("Minimum Sales for each year from 2021 to 2024:\n", min_sales_year)
max_sales_year = np.max(sales_data[:, 1:], axis=0)
print("Maximum Sales for each year from 2021 to 2024:\n", max_sales_year)
```

```python
# Avg sales for each company
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("Average Sales for each company from 2021 to 2024:\n", avg_sales)

# Avg sales for each year
avg_sales_year = np.mean(sales_data[:, 1:], axis=0)
print("Average Sales for each year from 2021 to 2024:\n", avg_sales_year)
```

```python
# cummulative sales for each company
cumulative_sales = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative Sales for each company from 2021 to 2024:\n", cumulative_sales)
```

```python
# Plotting the cumulative sales for each company

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumulative_sales, axis=0), marker='o', linestyle='-', color='b')
plt.title('Cumulative Sales for Each Company (2021-2024)')
plt.xlabel('Years (2021-2024)')
plt.ylabel('Cumulative Sales')
plt.grid(True)


plt.show()
```
