```python
import numpy as np
import matplotlib.pyplot as plt
```

Load and convertion of images

```python
array1 = np.array([1, 2, 3, 4, 5])

np.save('array1.npy', array1) # Save the array to a file named 'array1.npy'
```

```python
loaded_array = np.load('array1.npy') # Load the array from the file
print("Loaded Array:", loaded_array) # Print the loaded array
```

Let's try with some images(as every image is a data of RBG values)

```python
try:
    logo = np.load("numpy-logo.npy")

    # Display the logo using matplotlib
    plt.figure(figsize=(6, 6))
    plt.subplot(121)
    plt.title("NumPy Logo")
    plt.imshow(logo)

    dark_logo = 1 - logo  # Invert the colors for dark mode
    plt.figure(figsize=(6, 6))
    plt.subplot(122)
    plt.title("NumPy Logo (Dark Mode)")
    plt.imshow(dark_logo)


except Exception as e:
    print("An error occurred:", e)
```
